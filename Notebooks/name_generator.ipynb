```python
import names
import pandas as pd
import urllib.request
import random
```

```python
#Create random name and last names
word_url = "http://svnweb.freebsd.org/csrg/share/dict/words?view=co&content-type=text/plain"
response = urllib.request.urlopen(word_url)
long_txt = response.read().decode()
words = long_txt.splitlines()

upper_words = [word for word in words if word[0].isupper()]
name_words  = [word for word in upper_words if not word.isupper()]
one_name = ' '.join([upper_words[random.randint(0, len(upper_words))] for i in range(2)])


def rand_name():
   name = ' '.join([upper_words[random.randint(0, len(upper_words))] for i in range(2)])
   return name

random_name=[]
for n in range(100):
    name = rand_name()
    random_name.append(name)
    #print(name)
#random_name
sr = pd.Series(random_name)
sr
```

```python
# Create random ages
# random numbers and 
# append them to a list 
import random 
  
# Function to generate 
# and append them  
# start = starting range, 
# end = ending range 
# num = number of  
# elements needs to be appended 
def Rand(start, end, num): 
    res = [] 
  
    for j in range(num): 
        res.append(random.randint(start, end)) 
  
    return res 
  
# Driver Code 
num = 100
start = 20
end = 65
random_age=(Rand(start, end, num)) 
#random_age
sa = pd.Series(random_age)
sa
```
